```python
a = tf.constant(6.0)
b = tf.constant(3.0)


add = tf.add(a, b)
sub = tf.subtract(a, b)
mul = tf.multiply(a, b)
div = tf.divide(a, b)
pow_ = tf.pow(a, b)

print("Add:", add.numpy())
print("Sub:", sub.numpy())
print("Mul:", mul.numpy())
print("Div:", div.numpy())
print("Pow:", pow_.numpy())
```

```python
import tensorflow as tf

# Define 2D matrices
A = tf.constant([[1.0, 2.0],
                 [3.0, 4.0]])

B = tf.constant([[5.0, 6.0],
                 [7.0, 8.0]])

# Element-wise operations
add = tf.add(A, B)
sub = tf.subtract(A, B)
mul = tf.multiply(A, B)
div = tf.divide(A, B)
pow_ = tf.pow(A, B)

print("Add:\n", add.numpy())
print("Sub:\n", sub.numpy())
print("Mul:\n", mul.numpy())
print("Div:\n", div.numpy())
print("Pow:\n", pow_.numpy())
```
